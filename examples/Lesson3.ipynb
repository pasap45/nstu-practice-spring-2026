```python
x = X_train[:8].astype(np.float32)
y = y_train[:8]
```

```python
class NLLLoss:
    def forward(self, x: np.ndarray, y: np.ndarray) -> np.ndarray:
        batch_size = x.shape[0]
        self.grad = np.zeros_like(x)
        self.grad[np.arange(batch_size), y] = -1 / batch_size
        return -np.mean(x[np.arange(batch_size), y])

    def backward(self) -> np.ndarray:
        return self.grad
```

```python
numpy_loss_fn = NLLLoss()

numpy_res = numpy_model.forward(x)
loss = numpy_loss_fn.forward(numpy_res, y)
numpy_model.backward(numpy_loss_fn.backward())
print(loss)
```

```python
for _ in range(10):
    numpy_res = numpy_model.forward(x)
    loss = numpy_loss_fn.forward(numpy_res, y)
    numpy_model.backward(numpy_loss_fn.backward())
    print(loss)

    for p, g in zip(numpy_model.parameters, numpy_model.grad, strict=True):
        p += -1e-3 * g
```

```python
import optuna


def train(trial):
    hidden_layer_size = trial.suggest_int("hidden_layer_size", 1, 200)
    model = Exercise.create_model(
        Exercise.create_linear_layer(64, hidden_layer_size),
        Exercise.create_relu_layer(),
        Exercise.create_linear_layer(hidden_layer_size, 10),
        # Exercise.create_logsoftmax_layer(),
    )
    loss_fn = Exercise.create_cross_entropy_loss()

    lr = trial.suggest_float("learning_rate", 1e-5, 1e-2, log=True)
    n_epoch = trial.suggest_int("num_epoch", 1, 10)
    batch_size = trial.suggest_int("batch_size", 1, 100)

    Exercise.train_model(model, loss_fn, X_train, y_train, lr, n_epoch, batch_size)

    return (np.argmax(model.forward(X_val), axis=1) == y_val).mean()


study = optuna.create_study(direction="maximize")
study.optimize(train, n_trials=100)

study.best_value, study.best_params
```
